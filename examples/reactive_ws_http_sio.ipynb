```python
import tributary as t
```

```python
ws = t.WebSocket('wss://ws.paine.nyc', wrap=True, json=True)
p = t.Print(ws)
f = t.Perspective(p, schema={"A": float, "B": float, "C": float, "D": float}, plugin='y_line')
```

```python
t.run(f)
```

```python
t.GraphViz(f)
```

```python
http = t.HTTP('https://unpkg.com/@jpmorganchase/perspective-examples@0.1.18/build/citibike.json', json=True, field='stationBeanList')
psp = t.Perspective(http, schema={"id":int,
                                  "stationName": str,
                                  "availableDocks": int,
                                  "totalDocks": int,
                                  "latitude": float,
                                  "longitude": float,
                                  "statusValue": str,
                                  "statusKey":int,
                                  "availableBikes":int,
                                  "stAddress1": str,
                                  "stAddress2": str,
                                  "city": str,
                                  "postalCode": str,
                                  "location": str,
                                  "altitude": str,
                                  "testStation": bool,
                                  "lastCommunicationTime": "datetime",
                                  "landMark": str}, plugin='xy_scatter', columns=['longitude', 'latitude', 'availableBikes'])
```

```python
t.run(psp);
```

```python
t.GraphViz(psp, 'psp')
```
