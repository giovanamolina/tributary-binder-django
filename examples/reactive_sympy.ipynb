```python
import tributary.symbolic as ts
import tributary.reactive as tr
```

```python
expr = ts.parse_expression("10sin**2 x**2 + 3xyz + tan theta")
```

```python
clz = ts.construct_streaming(expr)

class Derived(clz):
    def __init__(self, *args, **kwargs):
        super(Derived, self).__init__(*args, **kwargs)

    def run(self):
        return tr.run(tr.Print(self._run))
```

```python
def foo(*args):
    for _ in range(5):
        yield _

x = Derived(x=tr.Const(1), y=tr.Foo(foo), z=tr.Timer(1, repeat=1), theta=tr.Const(4))
```

```python
x.graphviz()
```

```python
ret = x.run()
```
