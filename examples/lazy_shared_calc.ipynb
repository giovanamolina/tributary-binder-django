```python
import tributary.symbolic as ts
import tributary.lazy as tl
```

```python
expr = ts.parse_expression("10sin**2 x**2 + 3xyz + tan theta")
```

```python
clz = ts.construct_lazy(expr)

class Derived(clz):
    def __init__(self, **kwargs):
        super(Derived, self).__init__(**kwargs)

    @tl.node
    def out2(self):
        return self.evaluate() + 1

    @tl.node
    def out3(self):
        return self.evaluate() + self.out2()

x = Derived(x=1, y=2, z=3, theta=4)
```

```python
x.evaluate()()
```

```python
x.evaluate().graphviz()
```

```python
x.out2()()
```

```python
x.out2().graphviz()
```

```python
x.out3()()
```

```python
x.out3().graphviz()
```
