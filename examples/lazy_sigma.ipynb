```python
import tributary.symbolic as ts
expr = ts.parse_expression("10sin**2 x**2 + 3xyz + tan theta")
clz = ts.construct_lazy(expr)
x = clz(x=1, y=2, z=3, theta=4)
```

```python
x.evaluate()()
```

```python
x.x = 2
x.evaluate()()
```

```python
x.evaluate().graphviz()
```
